# Jupyter Notebook

De acordo com o site do projeto, o Jupyter Notebook é um aplicativo da Web de código aberto que permite criar e compartilhar documentos que contêm código ativo, equações, visualizações e texto narrativo.
 
Este tipo de aplicativo é um recurso que permite estudo interativo, onde o leitor pode executar e modificar código e ver os resultados sem sair da pagina que está lendo.

Sua linguagem padrão é o Python. Mas ele pode conectar-se a linguagens como o R, Julia, Ruby, Scala e Haskell. Atualmente, são suportadas mais de 40 linguagens de programação.

# Atalhos de teclado, para o Jupyter Notebook 

Esc – Ativa o modo de comando, onde você pode navegar saltar as células de seu notebook com as teclas de seta.

Enquanto  permanecer no modo de comando:

A - Para inserir uma nova célula acima da célula atual;

B - Para inserir uma nova célula abaixo;

M - Para alterar a célula atual para Markdown;

Y - Para alterá-lo de volta para o código;

D + D (pressione a tecla duas vezes) - Para excluir a célula atual;

Enter - O levará do modo de comando de volta ao modo de edição, onde você pode editar o texto nas células;

Shift + Tab -  Mostrara o Docstring (documentação) para o objeto que você acabou de digitar em uma célula de código;

Ctrl + Shift + - Dividi a célula atual em 2 (duas)  onde estiver o seu cursor;

Esc + F - Encontre e substitua seu código, mas não as saídas;

Esc + - O Alterna a saída da célula.

Selecione várias células:

Shift + J - Seleciona a próxima venda em uma direção para baixo; 

Shift + K - Seleciona vendas para cima;

Depois que as células são selecionadas, você pode excluir / copiar / recortar / colar / executá-las como um lote.

Shift + M- Mesclar várias células.

# Lista:

Para criarmos uma lista não ordenada, temos que pular uma linha depois do ultimo paragrafo e colocarmos um * (asterisco) a frente do item listado.  Além do * (asterisco), podemos utilizar o sinal de + (soma) ou – (subtração) pra formarmos uma lista.

veja:

*Azul
*Verde
*Amarelo
*Rosa

Pra termos uma lista ordenada, 

1.Azul
2.Verde
3.Amarelo
4.Rosa

Para criamos uma lista aninhada, basta uma colocarmos o subitem com um espaço na frente:

1.Azul
   1.Bola
  2.Triangulo
  3.Quadrado
2.Verde
3.Amarelo
